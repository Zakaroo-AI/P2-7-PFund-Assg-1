```python
ticker_dict = {}
for ticker in TICKERS:
    ticker_dict[ticker] = yf.Ticker(ticker)
```

```python
three_year_2022 = {}
for ticker in TICKERS:
    three_year_2022[ticker] = ticker_dict[ticker].history(period = '1y', start = '2022-01-01', interval = '1d')
    three_year_2022[ticker].to_csv(f'{ticker}_2022.csv')
    
```
