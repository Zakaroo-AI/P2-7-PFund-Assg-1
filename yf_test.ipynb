```python
import time

import pandas as pd
import yfinance as yf
```

```python
# TICKERS = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'JNJ', 'XOM', 'WMT', 'SPY', 'BTC-USD']
TICKERS = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']
```

```python
ticker_dict = {}
for ticker in TICKERS:
    ticker_dict[ticker] = yf.Ticker(ticker)
```

```python
ticker_hists = []
for ticker in TICKERS:
    ticker_hists.append(ticker_dict[ticker].history(period = '3y', interval = '1d'))
```
